```python
import numpy as np
```

```python
def ne_vzaimno(x,y):
    for i in range(2,min(x,y)//2+2):
        if(x%i==0 and y%i==0): return True
    return False
ne_vzaimno(12,4)
```

```python
def f11(a):
    v=np.array([])
    for i in range(2,a+1,2):
        if ne_vzaimno(i,a): v = np.append(v,i)
    return len(v)
a=int(input())
f11(a)
```
