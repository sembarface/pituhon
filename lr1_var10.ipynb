## Задание 1


```python
def nod(a,b):
    while b!=0:
        a,b=b,a%b
    return a

def f1_1(n):
    c=0
    i=2
    while i<=n:
        if nod(i,n)!=1:
            c=c+1
        i=i+2
    return c
```

```python
n=int(input())
print(f1_1(n))
```

```python
def f1_2(n):
    n=abs(n)
    m=-1
    if n==0:
        return -1
    while n>0:
        d=n%10
        if d%3!=0 and d>m:
            m=d
        n=n//10
    return m
```

```python
n=int(input())
print(f1_2(n))
```

```python
def min_del(n):
    n=abs(n)
    if n<2:
        return 1
    i=2
    while i<=n:
        if n%i==0:
            return i
        i=i+1
    return n

def sum_m5(n):
    n=abs(n)
    s=0
    if n==0:
        return 0
    while n>0:
        d=n%10
        if d<5:
            s=s+d
        n=n//10
    return s

def f1_3(n):
    n=abs(n)
    d=min_del(n)
    m=0
    i=1
    while i<=n:
        if nod(i,n)!=1 and i%d!=0:
            m=i
        i=i+1
    return m*sum_m5(n)
```

```python
n=int(input())
print(f1_3(n))
```

```python
def m11_1(s):
    gl='aeiouyауоыиэяюёеAEIOUYАУОЫИЭЯЮЁЕ'
    g=0
    c=0
    for ch in s:
        if ('a'<=ch<='z') or ('A'<=ch<='Z') or ('а'<=ch<='я') or ('А'<=ch<='Я') or ch=='ё' or ch=='Ё':
            if ch in gl:
                g=g+1
            else:
                c=c+1
    if g+c==0:
        return 0
    ag=g/(g+c)
    ac=c/(g+c)
    if ac>=ag:
        return ac-ag
    return ag-ac

def f11_1(a):
    return sorted(a,key=m11_1)
```

```python
n=int(input())
a=[]
for i in range(n):
    a.append(input())
r=f11_1(a)
for x in r:
    print(x)
```

```python
def m11_5(s):
    t=''
    for ch in s:
        if ch!=' ':
            t=t+ch
    if t=='':
        return 0
    m=0
    i=0
    while i<len(t):
        c=0
        j=0
        while j<len(t):
            if t[j]==t[i]:
                c=c+1
            j=j+1
        if c>m:
            m=c
        i=i+1
    p=m/len(t)
    e=1/26
    return (p-e)*(p-e)

def f11_5(a):
    return sorted(a,key=m11_5)
```

```python
n=int(input())
a=[]
for i in range(n):
    a.append(input())
r=f11_5(a)
for x in r:
    print(x)
```

```python
def m11_9(s):
    if s=='':
        return 0
    mx=ord(s[0])
    for ch in s:
        if ord(ch)>mx:
            mx=ord(ch)
    k=len(s)//2
    if k==0:
        d=0
    else:
        sm=0
        i=0
        while i<k:
            sm=sm+abs(ord(s[i])-ord(s[len(s)-1-i]))
            i=i+1
        d=sm/k
    return (mx-d)*(mx-d)

def f11_9(a):
    return sorted(a,key=m11_9)
```

```python
n=int(input())
a=[]
for i in range(n):
    a.append(input())
r=f11_9(a)
for x in r:
    print(x)
```

```python
def f15_10(a,b):
    c=0
    u=[]
    for x in a:
        if x in b and x not in u:
            u.append(x)
            c=c+1
    return c
```

```python
a=list(map(int,input().split()))
b=list(map(int,input().split()))
print(f15_10(a,b))
```

```python
def f15_22(a,l,r):
    t=[]
    for x in a:
        if x>=l and x<=r:
            t.append(x)
    if len(t)==0:
        return 0
    m=t[0]
    for x in t:
        if x<m:
            m=x
    c=0
    for x in t:
        if x==m:
            c=c+1
    return c
```

```python
a=list(map(int,input().split()))
l,r=map(int,input().split())
print(f15_22(a,l,r))
```

```python
def f15_46(a):
    p=[]
    n=[]
    z=[]
    for x in a:
        if x>0:
            p.append(x)
        elif x<0:
            n.append(x)
        else:
            z.append(x)
    return p+n+z
```

```python
def f15_58(a):
    c=0
    i=0
    while i<len(a):
        ok=False
        j=0
        while j<len(a) and not ok:
            if j!=i:
                k=j+1
                while k<len(a):
                    if k!=i and a[j]+a[k]==a[i]:
                        ok=True
                        break
                    k=k+1
            j=j+1
        if ok:
            c=c+1
        i=i+1
    return c
```
