## Задание 1


```python
n, k = map(int, input().split())
d ={}
for _ in range(k):
    a, b = map(int, input().split())
    for i in range(a,n+1,b):
        if not(i%7==0 or (i+1)%7==0):
            d[i]=1
print(sum(d.values()))
'''
Ввод Вывод
19 3  8
2 3
3 5
9 8
'''
d
```

## Задание 2


```python
with open('data_lab2.txt') as file:
    d={}
    for line in file:
        s = line.lower().split()
        if s[0] in d.keys():
            if s[1] in d[s[0]].keys():
                d[s[0]][s[1]] += int(s[2])
            else:
                d[s[0]][s[1]] = int(s[2])
        else:
            d[s[0]]={s[1]:int(s[2])}

d = {k: dict(sorted(v.items())) for k, v in d.items()}
d = dict(sorted(d.items()))
for i in d.items():
    print(f'{i[0].title()}:')
    for k, v in i[1].items():
        print(k, v)
```
